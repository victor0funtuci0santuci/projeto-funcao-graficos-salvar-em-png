# Projeto códigos automaticos para gerar e salvar Graficos


**Introdução**

Bom nesse Projeto iremos trabalhar com a base de nascidos de Roraima. Antes de tudo nós ja tratamos os dados e dividimos os nossos dados em arquivos por mês.

Nosso objetivo é criar um código para gerar gráficos referentes a base de dados de março, abril, maio, junho e dezembro. Supomos que alguem nós solicite no final de cada mês um relatorio de fechamento, para ver as informações de nascidos do mês, com isso vamos desenvolver um codigo onde ele gera esses graficos e após gerar ele vai salvalos em nossa area de trabalho.

## Março

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os

sns.set()  

def plota_pivot_table(df, value, index, func, ylabel, xlabel, opcao='nada'):
    if opcao == 'nada':
        pd.pivot_table(df, values=value, index=index,aggfunc=func).plot(figsize=[15, 5])
    elif opcao == 'unstack':
        pd.pivot_table(df, values=value, index=index,aggfunc=func).unstack().plot(figsize=[15, 5])
    elif opcao == 'sort':
        pd.pivot_table(df, values=value, index=index,aggfunc=func).sort_values(value).plot(figsize=[15, 5])
    plt.ylabel(ylabel)
    plt.xlabel(xlabel)
    return None

sinasc = pd.read_csv('SINASC_RO_2019_MAR.csv')


max_data = sinasc.DTNASC.max()[:7]
print(max_data)

# Criar o diretório se não existir
os.makedirs('./Graficos/'+max_data, exist_ok=True)

plota_pivot_table(sinasc, 'IDADEMAE', 'DTNASC', 'mean', 'média idade mãe por data', 'data nascimento')
plt.savefig('./Graficos/'+max_data+'/media idade mae por sexo.png')# essa ultima parte vai ser o nome do arquivo

plota_pivot_table(sinasc, 'IDADEMAE', ['DTNASC', 'SEXO'], 'mean', 'media idade mae','data de nascimento','unstack')
plt.savefig('./Graficos/'+max_data+'/media peso bebe por sexo.png')

plota_pivot_table(sinasc, 'PESO', ['DTNASC', 'SEXO'], 'mean', 'media peso bebe','data de nascimento','unstack')
plt.savefig('./Graficos/'+max_data+'/media apgar1 por escolaridade mae.png')

plota_pivot_table(sinasc, 'PESO', 'ESCMAE', 'median', 'PESO mediano','escolaridade mae','sort')
plt.savefig('./Graficos/'+max_data+'/media apgar1 por gestacao.png')

plota_pivot_table(sinasc, 'APGAR1', 'GESTACAO', 'mean', 'apgar1 medio','gestacao','sort')
plt.savefig('./Graficos/'+max_data+'/media apgar5 por gestacao.png')
```
